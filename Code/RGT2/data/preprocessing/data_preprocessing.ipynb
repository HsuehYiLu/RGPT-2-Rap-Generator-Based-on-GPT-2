```python
def getting_data(path):
	"""function to parse data

	Args:
		path (str): path to data file

	Returns:
		list[list[str]]: list of lists(verses)
	"""
	verses_all = []
	for _file in os.listdir(path):
		with open(os.path.join(path, _file), "r") as file:
			content = file.read()
			# split on songs
			songs = content.split("\n\n\n")
			# splilt on verses
			_verses = [song.split("\n\n") for song in songs]
			# split on bars
			verses = [bar.split("\n") for song in _verses for bar in song]
			verses_all.extend(verses)
	return verses_all
```

```python
verses_all = getting_data(path)
```

```python
def parse_song(verse):
    """Parse a full verse into a dictionary containing a list of lines

    Args:
        verse (list[str]): one verse

    Returns:
        dict[str, str]: format in dictionary
    """
    return {'verse': verse}

```

```python
def verses_to_json(verses, json_path):
    """Parse a list of verse strings and dump it as JSON

    Args:
        verses (list[list[str]]): list of lists(verses)
        json_path (str): output file name
    """
    output = {'count': 0, 'verses': {}}
    for index, verse in enumerate(verses):
        verse_dict = parse_song(verse)
        if verse_dict:
            output['verses'][index] = verse_dict

    output['count'] = len(output['verses'])

    with open(json_path, 'w') as outfile:
        json.dump(output, outfile)
```

```python
verses_to_json(verses_all, 'verses.json')
```
